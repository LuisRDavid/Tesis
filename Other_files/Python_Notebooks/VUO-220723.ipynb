## Importar las librarias necesarias

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

Importar los datasets 

```python
ds = pd.read_csv('../VUO-220723.csv')
ds
```

```python
time = ds['TIEMPO'].values.reshape(-1,1)
r1 = ds[' OD_R1'].values.reshape(-1,1)
r2 = ds[' OD_R2'].values.reshape(-1,1)
r3 = ds[' OD_R3'].values.reshape(-1,1)
r4 = ds[' OD_R4'].values.reshape(-1,1)
```

```python
fig = plt.figure()
ax = fig.subplots()
plt.plot(r1, time, color='blue', label='Reactor 1')
plt.scatter(r1, time, color='blue', edgecolors='gray')
plt.plot(r2, time, color='red', label='Reactor 2')
plt.scatter(r2, time, color='red', edgecolors='gray')
plt.plot(r3, time, color='green', label='Reactor 3')
plt.scatter(r3, time, color='green', edgecolors='gray')
plt.plot(r4, time, color='black', label='Reactor 4')
plt.scatter(r4, time, color='black', edgecolors='gray')
plt.legend()
plt.grid()
plt.show()
```
