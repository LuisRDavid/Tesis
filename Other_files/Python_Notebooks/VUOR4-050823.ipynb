```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('../VUO-050823.csv')
time = df['TIEMPO'].values.reshape(-1,1)
od = df['OD_R4'].values.reshape(-1,1)
```

```python
linearRegressor = LinearRegression()
linearRegressor.fit(time[3:], od[3:])
odPredict = linearRegressor.predict(time[3:])
odPredict
m = linearRegressor.coef_[0][0]
c = linearRegressor.intercept_[0]
label = r'$VUO= %0.4fx %+0.4f$' % (m, c)
print(label)
R2 = r2_score(time[3:], odPredict)
print(R2)
```

```python
fig = plt.figure()
ax = fig.subplots()
plt.plot(time[3:], od[3:])
plt.plot(time[3:], odPredict)
plt.grid(visible=True)
ax.text(2.5, 6.4, label)
ax.text(2.5, 6.35, r'$R^2= %0.4f$'%(R2))
```
