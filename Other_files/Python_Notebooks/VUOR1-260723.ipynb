```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('../VUO-260723.csv')
time = df['TIEMPO'].values.reshape(-1,1)
od = df['OD_R1'].values.reshape(-1,1)
```

```python
linearRegressor = LinearRegression()
linearRegressor.fit(time[:], od[:])
odPredict = linearRegressor.predict(time[:])
odPredict
m = linearRegressor.coef_[0][0]
c = linearRegressor.intercept_[0]
label = r'$VUO= %0.4fx %+0.4f$' % (m, c)
print(label)
R2 = r2_score(time[:], odPredict)
print(R2)
```

```python
fig = plt.figure()
ax = fig.subplots()
plt.plot(time[0:], od[0:])
plt.plot(time[0:], odPredict)
plt.grid(visible=True)
ax.text(2.5, 2, label)
ax.text(2.5, 1, r'$R^2= %0.4f$'%(R2))
```
