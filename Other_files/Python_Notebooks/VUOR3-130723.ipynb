```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv('../VUO-130723.csv')
time = df['TIEMPO'].values.reshape(-1,1)
od = df['OD_R3'].values.reshape(-1,1)

linearRegressor = LinearRegression()
linearRegressor.fit(time[4:20], od[4:20])
odPredict = linearRegressor.predict(time[4:20])
odPredict
m = linearRegressor.coef_[0][0]
c = linearRegressor.intercept_[0]
label = r'$VUO= %0.4fx %+0.4f$' % (m, c)
print(label)
R2 = r2_score(time[4:20], odPredict)
print(R2)

fig = plt.figure()
ax = fig.subplots()
plt.plot(time[4:20], od[4:20])
plt.plot(time[4:20], odPredict)
plt.grid(visible=True)
ax.text(4, 6.78, label)
ax.text(4, 6.77, r'$R^2= %0.4f$'%(R2))
```
